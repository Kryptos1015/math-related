```python
# Pr√©liminaires

from scipy.integrate import odeint
import numpy as np
import pylab as pl
import math
```

```python
# Exercise 2a (programme intitial)

N = 10
t = 2 * np.pi * np.linspace(0, 1, N)
print("t = ", t)
f = np.sin(t)
print("f = ", f)
```

```python
# Exercise 2b (programme initial)

pl.plot(t, f, "o-", label="Fonction sinus")
pl.xlabel("Variable t")
pl.ylabel("Fonction f(t)")
pl.legend(loc="best")
pl.show()
```

```python
# Exercise 2b (apr√©s modifications)

N = 100
t = np.linspace(-np.pi, np.pi, N)
f = np.sin(t)
g = np.cos(t)

pl.plot(t, f, "o-", label="Fonction sinus")
pl.plot(t, g, "o-", label="Fonction cosinus")
pl.xlabel("Variable t")
pl.ylabel("Fonctions f(t) et g(t)")
pl.legend(loc="best")
pl.show()
```

```python
# Exercise 3: Tracer la fonction f(x) avec Python

a, v, w = 1, 2, 3
fx = []
x_vars = np.linspace(0, 2, N)

# EQUATION INITALE (ENSUITE SIMPLIFI√â)
# (ALPHA * math.exp(a * x)) + (BETA * math.exp(-a * x)) + ((math.exp(w*x)) / (w**2 - a**2))

for x in x_vars:
    fx.append((math.exp(a*x) * (((-w - a) / ((2*a) * (w**2 - a**2))) + v)) + (math.exp(-a * x) * (((w-a) / ((2*a) * (w**2 - a**2))) - v)) + ((math.exp(w*x)) / (w**2 - a**2)))
```

```python
pl.plot(x_vars, fx, "o-")
pl.xlabel("Variable x")
pl.ylabel("f(x)")
pl.title("Trac√© de la fonction f(x)")
pl.show()
```

```python
# Exercise 4: R√©solution num√©rique d'une √©quation diff√©rentielle ordinaire

dy_dx = 0 # pour x = 0

def gx(y, x):
    return (a * math.exp(a*x) * (((-w - a) / ((2*a) * (w**2 - a**2))) + v)) + (-a * math.exp(-a * x) * (((w-a) / ((2*a) * (w**2 - a**2))) - v)) + ((w * (math.exp(w*x))) / (w**2 - a**2))

g_y = odeint(gx, dy_dx, x_vars)
pl.plot(x_vars, g_y)
pl.show()
```
