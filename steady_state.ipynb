```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(27)

x, y = [], []

A = [[0.8,0.1],[0.2,0.9]]
B = [[0],[1]]

x.append(np.matmul(A,B)[0][0])
y.append(np.matmul(A,B)[1][0])
print(x, y)
```

```python
for i in range(1,100):
    
    steady_state = 0
    
    An = A.copy()
    for n in range(i):
        An = np.matmul(An, A)
        
    x.append(np.matmul(An,B)[0][0])
    y.append(np.matmul(An,B)[1][0]) 
    
    if steady_state == 0:
        if y[-1] == y[-2]:
            steady_state = y[-1]
```

```python
steady_state_index = y.index(steady_state)
print(steady_state)
print(f"The power at which A^n reaches a steady state is {steady_state_index}.")
```

```python
plt.plot(y)
plt.title("Long term probabilities")
plt.xlabel("n (Power)")
plt.ylabel("Probability")
plt.plot(0.66)
plt.show()
```
